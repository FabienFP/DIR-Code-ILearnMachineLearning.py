```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
# Importing the dataset
dataset = pd.read_csv('../00_Datasets/StackOverflow/StackOverflow.csv', header = 0)
```

```python
dataset.info()
```

<b>NOTE:</b> As you can see the data set contains over 150 columns and 51k rows. After a short analysis I decided to focus on the questions that interested me the most. Let's create a list with column names for these questions:

```python
interesting_columns = [
    'Professional',
    'Country',
    'EmploymentStatus',
    'HomeRemote',
    'DeveloperType',
    'YearsProgram',
    'JobSatisfaction',
    'PronounceGIF',
    'ClickyKeys',
    'TabsSpaces',
    'EducationTypes',
    'HaveWorkedLanguage', 
    'WantWorkLanguage',
    'HaveWorkedFramework',
    'WantWorkFramework',
    'IDE',
    'VersionControl'
]
data = pd.DataFrame(dataset, columns=interesting_columns)
```

```python
data.info()
```

```python
plt.figure(figsize=(16,8))
sns.set_style("whitegrid")
plt.title('How do you pronounce "GIF"?', fontsize=25, fontweight='bold', y=1.05,)
sns.countplot(y="PronounceGIF", data=data, palette=sns.color_palette("Reds_d"));
plt.xlabel('', fontsize=15)
plt.ylabel('', fontsize=15)
plt.show()
```

```python
import operator
from collections import OrderedDict
languages = {}
for index, row in data.iterrows():
    if (row['HaveWorkedLanguage'] != None and type(row['HaveWorkedLanguage']) != float):
        languages_list = row['HaveWorkedLanguage'].split('; ')
        for lang in languages_list:
            if lang in languages.keys():
                languages[lang] += 1
            else:
                languages[lang] = 1
                
languages = OrderedDict(sorted(languages.items(), key=operator.itemgetter(1), reverse=True))
```

```python
plt.figure(figsize=(16,8))
sns.set_style("whitegrid")
plt.title('Most Popular Programming Languages', fontsize=20, fontweight='bold', y=1.05,)
sns.set(font_scale = 2)
b = sns.barplot(x=list(languages.keys()), y=list(languages.values()), palette="hls")
plt.xticks(rotation=90)
b.tick_params(labelsize=15)
plt.xlabel("Programming Languages", fontsize=20)
plt.ylabel('Number of developers', fontsize=20)
plt.show()
```
